__Importing all necessary libraries like pandas, numpy, matplotlib, seaborn & scikit learn__

__Spliting data into two parts: feature and target.__

_like here we give values to the species i.e. Iris-setosa=0,Iris-versicolor=1,Iris-virginica=2_

```python
i_data = pd.DataFrame(iris_set.data, columns=iris_set.feature_names)
i_data["species"] = iris_set.target

fea_data = iris_set.data
tar_data = iris_set.target
```

__Spliting data into train and test data set.__

_Here I split the data into train and test data i.e. 20% data is used for testing and other is used for train the model._

```python
train, test = train_test_split(i_data, test_size = 0.2)
```

__Model Training__

```python
X_train, X_test, y_train, y_test = train_test_split(fea_data, tar_data, test_size=0.2, random_state=50)
print('Training successfull')
```

__Fitting data and creating object of Decision tree classifier.__

_Like other machine learning algorithm, decision tree also have two models: regression and classification. Here I implemented classification model as there are some discrete categorical target values. Here I fitted the data and created object of decision tree classifier._

```python
clf=DecisionTreeClassifier()
clf.fit(X_train,y_train)

print('Instance of an decision tree successfully created')
```

_Now I print the length of train data and test data._

```python
len(X_train)
```

```python
len(y_test)
```

__Checking the predicted value of train data.__

```python
y_pred = clf.predict(X_train)
print(y_pred)
```

__Checking the accuracy of prediction and confusion matrix__

_Here I check the accuracy of prediction done before._
_I used confusion matrix which shows that our predicted value is accurate based on train dataset and test dataset._

```python
print("Accuracy of trained data:\n",accuracy_score(y_train,y_pred))

plot_confusion_matrix(clf, X_test, y_test)  
plt.show()
```

__More Predictable version  of decision tree__

_Here is a more  predictible version of decision tree. Root node, decision node and leaf node........_

_How decision tree  works is , it starts with selection of best attribute based on split criteria then attribute is turned into decision node which breaks dataset into smaller subsets... this is how it starts to build desicion tree by repeating process until no more attribute remains._

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(clf,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('imagename.png')
```
